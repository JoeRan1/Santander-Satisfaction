```python
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd

class CustomTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        # Columns to drop
        self.drop_cols = [
            'saldo_medio_var29_ult1','saldo_medio_var29_ult3',
            'saldo_medio_var13_medio_ult3','saldo_medio_var13_medio_ult1',
            'saldo_medio_var13_medio_hace2','num_venta_var44_hace3',
            'num_trasp_var33_in_ult1','num_trasp_var17_in_hace3',
            'saldo_var6','saldo_var13_medio','saldo_var18'
        ]

        # Columns to binarize (1 if >0 else 0)
        self.bin_cols = [
            'imp_ent_var16_ult1','imp_op_var39_comer_ult1',
            'imp_op_var40_comer_ult1','imp_op_var40_comer_ult3',
            'imp_op_var40_efect_ult1','imp_op_var41_efect_ult3',
            'num_var1_0','num_var1','num_var12_0','num_var12',
            'num_var13_corto_0','num_var13_corto','num_var13_largo_0',
            'num_var13_largo','num_var13','num_var14_0','num_var14',
            'num_var17_0','num_var17','num_var24_0','num_var24',
            'num_var26_0','num_var26','num_var25_0','num_var25',
            'num_op_var40_hace2','num_op_var40_hace3','num_op_var40_ult1',
            'num_op_var40_ult3','num_op_var41_hace2','num_op_var41_hace3',
            'num_op_var41_ult1','num_op_var41_ult3','num_op_var39_hace2',
            'num_op_var39_hace3','num_op_var39_ult1','num_op_var39_ult3',
            'num_var5_0','num_var5','num_var8_0','num_var13_0',
            'num_var30_0','num_var30','num_var31_0','num_var31',
            'num_var32_0','num_var32','num_var33_0','num_var33',
            'num_var35','num_var37_med_ult2','num_var37_0','num_var37',
            'num_var39_0','num_var40_0','num_var41_0','num_var42_0','num_var42',
            'num_var44_0','saldo_var8','saldo_var12','saldo_var13_corto',
            'saldo_var13_largo','saldo_var20','saldo_var24','saldo_var26',
            'saldo_var25','saldo_var29','saldo_var30','saldo_var31',
            'saldo_var32','saldo_var44','saldo_var33','saldo_var34',
            'saldo_var37','saldo_var40','saldo_var42'
        ]

        # Columns with ternary mapping (1 if >0, -1 if <0, else 0)
        self.ternary_cols = [
            'delta_imp_aport_var13_1y3','delta_imp_aport_var17_1y3',
            'delta_imp_aport_var33_1y3','delta_imp_compra_var44_1y3',
            'delta_imp_reemb_var17_1y3','delta_imp_trasp_var17_in_1y3',
            'delta_imp_trasp_var33_in_1y3','delta_imp_venta_var44_1y3',
            'delta_num_aport_var13_1y3','delta_num_aport_var17_1y3',
            'delta_num_aport_var33_1y3','delta_num_compra_var44_1y3',
            'delta_num_reemb_var17_1y3','delta_num_trasp_var17_in_1y3',
            'delta_num_trasp_var33_in_1y3','delta_num_venta_var44_1y3',
            'imp_amort_var18_ult1','imp_amort_var34_ult1',
            'imp_aport_var13_hace3','imp_aport_var13_ult1',
            'imp_aport_var17_hace3','imp_aport_var17_ult1',
            'imp_aport_var33_hace3','imp_aport_var33_ult1',
            'imp_var7_emit_ult1','imp_var7_recib_ult1',
            'imp_compra_var44_hace3','imp_compra_var44_ult1',
            'imp_reemb_var13_ult1','imp_reemb_var17_ult1',
            'imp_var43_emit_ult1','imp_trans_var37_ult1',
            'imp_trasp_var17_in_hace3','imp_trasp_var17_in_ult1',
            'imp_trasp_var17_out_ult1','imp_trasp_var33_in_hace3',
            'imp_trasp_var33_in_ult1'
        ]

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()

        # 1. Drop columns
        X = X.drop(columns=self.drop_cols, errors='ignore')

        # 2. var3 special rule
        X['var3'] = X['var3'].apply(lambda x: 0 if x < 2 else 3 if x > 2 else x)

        # 3. Binary transforms
        for col in self.bin_cols:
            if col in X:
                X[col] = X[col].apply(lambda x: 1 if x > 0 else 0)

        # 4. var36 mapping
        if 'var36' in X:
            X['var36'] = X['var36'].map({99: 3, 3: 2, 1: 2, 2: 2, 0: 0})

        # 5. Ternary transforms
        for col in self.ternary_cols:
            if col in X:
                X[col] = X[col].apply(lambda x: 1 if x > 0 else -1 if x < 0 else 0)

        return X
```

### Meta Classifier

```python
# Imports
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier

# Assuming your individual pipelines are already defined:
# gb_model, lgbm_model, cat_model

# Define base estimators for stacking
estimators = [
    ('gb', gb_model),
    ('lgbm', lgbm_model),
    ('cat', cat_model)
]

# Define stacking classifier with Logistic Regression as meta-model
stack_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=500),
    cv=3,          # internal CV for stacking
    n_jobs=-1,
    passthrough=False   # use predictions from base models only
)

# Stratified K-Fold for cross-validation
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Cross-validate the stacking model
stack_scores = cross_val_score(
    stack_model,
    X,
    y,
    cv=skf,
    scoring='roc_auc',
    n_jobs=-1
)

print("Stacked ROC-AUC scores:", stack_scores)
print("Mean Stacked ROC-AUC:", stack_scores.mean())
```
